```python
import matplotlib.pyplot as plt
import numpy as np

def compute_sorori_shinzaemon(n_days=100):
    """Calculate the number of rice grains in Sorori Shinzaemon's anecdote.

    Parameters
    -----------
    n_days: int
        Number of days to receive rice (default: 100)

    Returns
    -------
    list_n_grains : list
        List of rice grains received each day
    list_total_grains : list
        List of cumulative rice grains received up to each day
    """

    list_n_grains = []
    list_total_grains = []
    total_grains = 0

    for day in range(1, n_days + 1):
        # Rice grains on current day: 2^(day-1)
        grains_today = 2 ** (day - 1)
        total_grains += grains_today

        list_n_grains.append(grains_today)
        list_total_grains.append(total_grains)

    return list_n_grains, list_total_grains

def plot_rice_growth(n_days=100):
    """Plot the daily and cumulative rice grains over time."""

    list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days)
    days = list(range(1, n_days + 1))

    plt.figure(figsize=(12, 6))

    # Plot daily rice grains
    plt.subplot(1, 2, 1)
    plt.plot(days, list_n_grains, 'b-', linewidth=2)
    plt.yscale('log')  # Use logarithmic scale for better visualization
    plt.xlabel('Days')
    plt.ylabel('Rice Grains (log scale)')
    plt.title('Daily Rice Grains Received')
    plt.grid(True, alpha=0.3)

    # Plot cumulative rice grains
    plt.subplot(1, 2, 2)
    plt.plot(days, list_total_grains, 'r-', linewidth=2)
    plt.yscale('log')  # Use logarithmic scale
    plt.xlabel('Days')
    plt.ylabel('Cumulative Rice Grains (log scale)')
    plt.title('Cumulative Rice Grains Received')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    return list_n_grains, list_total_grains

def estimate_survival_days(total_grains, people=1, grains_per_meal=1000, meals_per_day=3):
    """Estimate how many days people can survive on given rice grains.

    Parameters
    -----------
    total_grains : int
        Total number of rice grains available
    people : int
        Number of people to feed (default: 1)
    grains_per_meal : int
        Rice grains needed per meal (default: 1000)
    meals_per_day : int
        Meals per day per person (default: 3)

    Returns
    -------
    survival_days : float
        Number of days the rice can feed the people
    """

    # Calculate daily consumption
    daily_consumption = people * grains_per_meal * meals_per_day

    # Calculate survival days
    survival_days = total_grains / daily_consumption

    return survival_days

# Calculate for 100 days
list_n_grains, list_total_grains = compute_sorori_shinzaemon(100)

# Results for 100 days
print("=== Results for 100 Days ===")
print(f"Rice grains on day 100: {list_n_grains[-1]:,}")
print(f"Total rice grains after 100 days: {list_total_grains[-1]:,}")

# Compare different durations
print("\n=== Comparison of Different Durations ===")
for days in [30, 60, 81, 100]:
    _, total_grains = compute_sorori_shinzaemon(days)
    print(f"After {days} days: {total_grains[-1]:,} grains")

# Estimate survival time
print("\n=== Survival Estimation ===")
total_grains_100_days = list_total_grains[-1]

# Research shows about 1000 grains  0.03kg of rice
# An average person needs about 0.2kg rice per day (600-700 calories)
survival_days = estimate_survival_days(total_grains_100_days, people=1)
survival_years = survival_days / 365

print(f"One person can survive for: {survival_days:,.0f} days ({survival_years:,.1f} years)")

# For a family of 4
survival_days_family = estimate_survival_days(total_grains_100_days, people=4)
survival_years_family = survival_days_family / 365
print(f"Family of 4 can survive for: {survival_days_family:,.0f} days ({survival_years_family:,.1f} years)")

# Create visualizations
plot_rice_growth(100)
```

```python
import numpy as np

# Setting the seed for reproducibility
np.random.seed(0)

# Defining the mean and covariance matrix
mean = [-3, 0]
cov = [[1.0, 0.8],
       [0.8, 1.0]]

# Generating 500 samples from the multivariate normal distribution
samples = np.random.multivariate_normal(mean, cov, size=500)

print("Shape of samples:", samples.shape)
print(samples)
```

Visualizing by scatter plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title("2D Normal Distribution Samples")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.grid(True)
plt.show()
```

Visualizing by Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]

data = np.random.multivariate_normal(mean, cov, 500)
x = data[:, 0]
y = data[:, 1]

xmin = min(x.min(), y.min())
xmax = max(x.max(), y.max())

# Plot histograms
plt.figure(figsize=(12, 5))

# Histogram for x
plt.subplot(1, 2, 1)
plt.hist(x, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of X')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xlim(xmin, xmax)

# Histogram for y
plt.subplot(1, 2, 2)
plt.hist(y, bins=30, color='salmon', edgecolor='black')
plt.title('Histogram of Y')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xlim(xmin, xmax)

# Final layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style="whitegrid")

# Problem 1
np.random.seed(0)
mean1 = [-3, 0]
cov1 = [[1.0, 0.8], [0.8, 1.0]]

# Problem 4
np.random.seed(1)
mean2 = [0, -3]
cov2 = [[1.0, 0.8], [0.8, 1.0]]

data1 = np.random.multivariate_normal(mean1, cov1, 500)
data2 = np.random.multivariate_normal(mean2, cov2, 500)

# Create labels and combine data
labels1 = np.zeros(500)  # Label for Problem 1
labels2 = np.ones(500)   # Label for Problem 4

data_combined = np.vstack((data1, data2))
labels_combined = np.hstack((labels1, labels2))

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_combined[:, 0], y=data_combined[:, 1],
                hue=labels_combined, palette='viridis', legend='full')

plt.title('Scatter Plot of Two 2D Normal Distributions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Dataset Label', labels=['0', '1'])
plt.tight_layout()

plt.show()
```

Data Combination

```python
# Combining the two arrays vertically
combined_data_vstack = np.vstack((data1, data2))

print("Shape of combined_data_vstack:", combined_data_vstack.shape)
print("\nFirst 5 rows of the combined data:\n", combined_data_vstack[:5])
print("\nLast 5 rows of the combined data:\n", combined_data_vstack[-5:])
```

Another way of combining the data

```python
# Concatenating the arrays along axis 0 (rows)
combined_data_concat = np.concatenate((data1, data2), axis=0)

print("Shape of combined_data_concat:", combined_data_concat.shape)
print("\nFirst 5 rows of the combined data:\n", combined_data_concat[:5])
print("\nLast 5 rows of the combined data:\n", combined_data_concat[-5:])
```

```python
import numpy as np

# Data from "Problem 1"
np.random.seed(0)
mean1 = [-3, 0]
cov1 = [[1.0, 0.8], [0.8, 1.0]]
data1 = np.random.multivariate_normal(mean1, cov1, 500)

# Data from "Problem 4"
np.random.seed(1)
mean2 = [0, -3]
cov2 = [[1.0, 0.8], [0.8, 1.0]]
data2 = np.random.multivariate_normal(mean2, cov2, 500)

combined_data = np.vstack((data1, data2))

# Creating an array of 500 zeros for the data from Problem 1
labels_for_data1 = np.zeros((500, 1))

# Creating an array of 500 ones for the data from Problem 4
labels_for_data2 = np.ones((500, 1))

# Combining the labels vertically into a single (1000, 1) array
labels = np.vstack((labels_for_data1, labels_for_data2))

final_dataset = np.hstack((combined_data, labels))

print("Shape of the final dataset:", final_dataset.shape)

# Print the first and last few rows to show the labels are correctly attached
print("\nFirst 5 rows (labeled 0):")
print(final_dataset[:5])

print("\nLast 5 rows (labeled 1):")
print(final_dataset[-5:])
```

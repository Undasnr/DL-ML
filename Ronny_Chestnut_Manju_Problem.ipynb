```python
# Variables
initial_volume = 0.1 # in cubic meters (example volume of one chestnut bun)
target_volume = 1.41e36 # approx volume of solar system in cubic meters
doubling_interval = 5 # minutes

# Program Logic
volume = initial_volume
time = 0

while volume < target_volume:
  volume *= 2
  time += doubling_interval

print(f"It takes {time} minutes to fill the solar system with chestnut buns!")

# Making the Program a general-purpose function
def time_to_fill(initial_volume, target_volume, interval_minutes):
  time = 0
  volume = initial_volume

  while volume < target_volume:
    volume *= 2
    time += interval_minutes

  return time

# Example: Tokyo Dome (~1.24 million cubic meters)
dome_time = time_to_fill(0.1, 1.24e6, 5)
print(f"Time to fill Tokyo Dome: {dome_time} minutes")

# Visualizing the growth
import matplotlib.pyplot as plt

volumes = []
times = []
volume = initial_volume
time = 0

while volume < target_volume:
  volumes.append(volume)
  times.append(time)
  volume *= 2
  time += doubling_interval

plt.plot(times, volumes)
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (cubic meter)")
plt.title("Exponential Chestnut Bun Growth")
plt.grid(True)
plt.show()
```

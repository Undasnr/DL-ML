**Linear function**

```python
import numpy as np

# Creating x values from -50 to 50 at 0.1 intervals
x = np.arange(-50, 50.1, 0.1)  # Include 50 by going slightly beyond

# Computing y = 12x + 1
y = 12 * x + 1

print("x =", x)
print("y =", y)
```

**Array combination**

```python
import numpy as np

# Creating x and y
x = np.arange(-50, 50.1, 0.1)
y = 12 * x + 1

# Combining x and y into a (1001, 2) array
xy_combined = np.column_stack((x, y))

# Checking shape and first few rows
print("Shape:", xy_combined.shape)
print("First 5 rows:\n", xy_combined[:5])
print("Last 5 rows:\n", xy_combined[:-5])
```

**Find the gradient**

```python
import numpy as np

# Recreating x and y if needed
x = np.arange(-50, 50.1, 0.1)
y = 12 * x + 1

# Calculating gradient using finite differences
dy = np.diff(y)
dx = np.diff(x)
gradient = dy / dx

# Checking shape and first few values
print("Gradient shape:", gradient.shape)
print("First 5 gradients:", gradient[:5])
```

**Drawing a graph**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating x values
x = np.arange(-50, 50.1, 0.1)

# Computing y values for the linear function y = 12x + 1
y = 12 * x + 1

# Computing gradient using finite differences
gradient = np.gradient(y, x)

# Plotting the linear function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = 12x + 1', color='blue')
plt.title('Linear Function: y = 12x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()
plt.show()

# Plotting the gradient
plt.figure(figsize=(10, 6))
plt.plot(x, gradient, label='Gradient dy/dx', color='green')
plt.title('Gradient of y = 12x + 1 using Finite Differences')
plt.xlabel('x')
plt.ylabel('Gradient (dy/dx)')
plt.legend()
plt.tight_layout()
plt.show()
```

Python functionalization

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(func, x_range):
    x = np.arange(*x_range, 0.1)
    y = func(x)
    gradient = np.gradient(y, 0.1)
    return x, y, gradient

def plot_function_and_gradient(x, y, gradient, title):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Function', color='blue')
    plt.plot(x, gradient, label='Gradient', color='orange')
    plt.title(title, fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y / Gradient', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Equation 1: y = x^2
func1 = lambda x: x**2
x1, y1, grad1 = compute_gradient(func1, (-50, 50.1))
plot_function_and_gradient(x1, y1, grad1, 'Gradient of y = x^2')

# Equation 2: y = 2x^2 + 2x
func2 = lambda x: 2*x**2 + 2*x
x2, y2, grad2 = compute_gradient(func2, (-50, 50.1))
plot_function_and_gradient(x2, y2, grad2, 'Gradient of y = 2x^2 + 2x')

# Equation 3: y = sin(x/12)
func3 = lambda x: np.sin(x/12)
x3, y3, grad3 = compute_gradient(func3, (0, 50.1))
plot_function_and_gradient(x3, y3, grad3, 'Gradient of y = sin(x/12)')

print("Equation 1 Gradient (y = x^2):", grad1)
print("Equation 2 Gradient (y = 2x^2 + 2x):", grad2)
print("Equation 3 Gradient (y = sin(x/12)):", grad3)
```

**Find the minimum value**

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(func, x_range):
    x = np.arange(*x_range, 0.1)
    y = func(x)
    gradient = np.gradient(y, 0.1)
    return x, y, gradient

def analyze_minimum(x, y, gradient, title):
    min_index = y.argmin()
    min_value = y[min_index]
    min_x = x[min_index]

    slope_before = gradient[min_index - 1] if min_index > 0 else None
    slope_after = gradient[min_index + 1] if min_index < len(gradient) - 1 else None

    print(f" {title}")
    print(f"Minimum y value: {min_value:.4f} at x = {min_x:.4f}")

    if slope_before is not None:
        print(f"Slope before minimum: {slope_before:.4f}")
    else:
        print("Slope before minimum: N/A")

    if slope_after is not None:
        print(f"Slope after minimum: {slope_after:.4f}")
    else:
        print("Slope after minimum: N/A")

    print("-" * 50)

# Equation 1: y = x^2
func1 = lambda x: x**2
x1, y1, grad1 = compute_gradient(func1, (-50, 50.1))
analyze_minimum(x1, y1, grad1, 'y = x^2')

# Equation 2: y = 2x^2 + 2x
func2 = lambda x: 2*x**2 + 2*x
x2, y2, grad2 = compute_gradient(func2, (-50, 50.1))
analyze_minimum(x2, y2, grad2, 'y = 2x^2 + 2x')

# Equation 3: y = sin(x/12)
func3 = lambda x: np.sin(x/12)
x3, y3, grad3 = compute_gradient(func3, (0, 50.1))
analyze_minimum(x3, y3, grad3, 'y = sin(x/12)')
```

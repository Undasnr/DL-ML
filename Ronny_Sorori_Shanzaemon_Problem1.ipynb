**The Sorori Shinzaemon Problem**

```python
# Figuring out rice grains day by day

def compute_sorori_shinzaemon(n_days=100):

  # Creating a list to store the grains for each day. The list starts empty.
  list_n_grains = []

  # Creating a list for storing the total rice grains up to each day. The list starts empty
  list_total_grains = []

  current_day_grains = 1
  total_grains_so_far = 1
  list_n_grains.append(current_day_grains)
  list_total_grains.append(total_grains_so_far)

  for day_number in range(1, n_days):
    current_day_grains = current_day_grains * 2
    total_grains_so_far = total_grains_so_far + current_day_grains
    list_n_grains.append(current_day_grains)
    list_total_grains.append(total_grains_so_far)

  return list_n_grains, list_total_grains

list_n_grains_10, list_total_grains_10 = compute_sorori_shinzaemon(n_days=10)

print(f"Grains on each day (for 10 days): {list_n_grains_10}")
print(f"Total grains up to each day (for 10 days): {list_total_grains_10}")
```

**How many people can live on the total rice grains after 100 days**

```python
def compute_sorori_shinzaemon(n_days=100):
  list_n_grains = []
  list_total_grains = []

  current_day_grains = 1
  total_grains_so_far = 1

  list_n_grains.append(current_day_grains)
  list_total_grains.append(total_grains_so_far)

  for day_number in range(1, n_days):
    current_day_grains = current_day_grains * 2
    total_grains_so_far = total_grains_so_far + current_day_grains
    list_n_grains.append(current_day_grains)
    list_total_grains.append(total_grains_so_far)

  return list_n_grains, list_total_grains

def estimate_survival_duration(total_grains_received, grains_per_person_per_day):
  """
  Estimates how many people can be fed for 100 days given a total amount of rice and daily and daily rice per person.

  Parameters

  total_grains_received : int
    The total number of rice grains received.
  grains_per_person_day : int
    The number of rice grains one person needs per day.

  Returns

  num_people : float
    The estimated number of people who can survive for 100 days.

  """

  grains_needed_per_person_for_100_days = grains_per_person_per_day * 100
  num_people = total_grains_received / grains_needed_per_person_for_100_days

  return num_people

# Getting total rice grains after 100 days from Shinzaemon
_, list_total_grains_100 = compute_sorori_shinzaemon(n_days=100)
total_rice_shinzaemon = list_total_grains_100[-1]

grains_per_person_per_day = 40000

number_of_people = estimate_survival_duration(total_rice_shinzaemon, grains_per_person_per_day)

print(f"Total grains after 100 days: {total_rice_shinzaemon}")
print(f"With {grains_per_person_per_day} grains per person per day,")
print(f"Approximately {number_of_people:.2f} people can be fed for 100 days!")
```
